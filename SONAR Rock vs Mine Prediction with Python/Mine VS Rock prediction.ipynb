## Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SONAR Rock vs Mine Prediction with Python/Copy of sonar data.csv",header=None)
data.head()
```

```python
data[60].value_counts()
```

R -> ROCK

M -> MINE

```python
data.groupby(60).mean()
```

separating data and labels

```python
x = data.drop(columns=60,axis=1)
y = data[60]
```

```python
print(x)
```

```python
print(y)
```

training and test data

```python
x_trains,x_test,y_trains,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_trains.shape,x_test.shape)
```

MODEL TRAINING

```python
print(x_trains)
```

```python
print(y_trains)
```

```python
model1 = LogisticRegression()
```

```python
model1.fit(x_trains,y_trains)
```

Accuracy on training data

```python
x_trains_prediction = model1.predict(x_trains)
training_data_accuracy = accuracy_score(x_trains_prediction,y_trains)
```

```python
print(f"accuracy on training data: {training_data_accuracy}")
```

```python
x_test_prediction = model1.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print(f"accuracy on training data: {test_data_accuracy}")
```
